```python
import os

#path = input()
path = r".\alldocs"

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
            files.append(os.path.join(r, file))
#print(files)
```

# Creating Different Dictionaries

```python
termids=dict()
docids=dict()
term_index=dict()
term_index_list=[]
```

# Converting Words To Lower Case

```python
def convert_lower_case(text):
     return text.lower()
```

# Library For Tokenizing Words

```python
import nltk
def split_words(text):
     return nltk.word_tokenize(text)
```

# Checking Whether A Tokenize Word Have Appeared Before

```python
def check_termid(text):
    if text in termids:
        return True
    else:
        return False
```

# Stemming Tokenize Words Using Porter Stemmer

```python
from nltk.stem.porter import *
def apply_stemming(tokens):
    stemmer = PorterStemmer()
    return stemmer.stem(tokens)
```

# Adding Values in DocIDs Dictionary

```python
def add_to_docids(pathName,docid):
    docName = os.path.basename(pathName)
    docids[docName] = docid
```

# Adding Values in TermIDs Dictionary

```python
def add_to_termids(word,termid):
    termids[word]=termid
```

# Checking Previous Occurences of the Tokenize Words

```python
def get_termid(word):
    return termids.get(word)
```

# Adding Words to TermIndex

```python
def add_to_term_index_list(termid,docid,termPosition):
    term_index_values = [termid,1,1,docid,termPosition]
    term_index_list.append(term_index_values)
def add_to_term_index(termid,docid,termPosition):
    term_index_values = [1,1,docid,termPosition]
    term_index[termid]=term_index_values
def update_term_index(termid,docid,termPosition):
    term_index_values = term_index.get(termid)
    length=len(term_index_values)
    if docid-term_index_values[length-2]!=0:
        term_index_values[1]=term_index_values[1]+1
    term_index_values[0]=term_index_values[0]+1
    values_update=[docid,termPosition]
    result = term_index_values+values_update
    new_dict = dict()
    new_dict[termid]=result
    term_index.update(new_dict)
```

# A For Loop For Parsing Each File To Extract Relevant Information

To Create Docids dictionary,Termids dictionay and TermIndex

```python
import nltk
import re
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
tokenizer = RegexpTokenizer(r'\w+')
termindex = 1
docindex = 1
termposition=0
pattern = re.compile(r'\b(' + r'|'.join(stopwords.words('english')) + r')\b\s*') # Removing Stop Words
for txt in files:
    termposition=0
    add_to_docids(txt,docindex)
    docindex = docindex+1
    contents = ""
    with open(txt,'r', encoding='utf8', errors='ignore') as f:
        for line in f.readlines():
            contents += line
        contents_lower_case = convert_lower_case(contents)
        contents_without_numbers = re.sub(r'\d+', '', contents_lower_case) # Removing Numbers
        contents_without_stop_words = pattern.sub('', contents_without_numbers)
        tokens = tokenizer.tokenize(contents_without_stop_words)
        print(docindex-1,len(tokens))
        for word in tokens:
            termposition=termposition+1
            if word.isalpha():
                strip_word=word.strip()
                if not strip_word.isspace():
                    if len(strip_word)>1:
                        word_stemming = apply_stemming(strip_word)
                        if check_termid(word_stemming):
                          termid=get_termid(word_stemming)
                          add_to_term_index_list(termid,docindex-1,termposition)
                          update_term_index(termid,docindex-1,termposition)
                        else:
                          add_to_term_index_list(termindex,docindex-1,termposition)
                          add_to_term_index(termindex,docindex-1,termposition)
                          add_to_termids(word_stemming,termindex)
                          termindex=termindex+1
```

# MergeSort Function For Creating Sorted Term, Doc ID pair

```python
def mergeSort(temp): 
    if len(temp) >1: 
        mid = len(temp)//2 
        L = temp[:mid] 
        R = temp[mid:] 
  
        mergeSort(L) 
        mergeSort(R) 
  
        i = j = k = 0
          
        while i < len(L) and j < len(R):
            L_values=L[i]
            R_values=R[j]
            if L_values[0] < R_values[0]: 
                temp[k] = L[i] 
                i+=1
            elif L_values[0]>R_values[0]: 
                temp[k] = R[j] 
                j+=1
            elif L_values[0]==R_values[0]:
                if L_values[1]<R_values[1]:
                    temp[k] = L[i] 
                    i+=1
                elif L_values[1]>R_values[1]:
                    temp[k] = R[j] 
                    j+=1
                elif L_values[1]==R_values[1]:
                    if L_values[2]<R_values[2]:
                        temp[k] = L[i] 
                        i+=1
                    elif L_values[2]>=R_values[2]:
                        temp[k] = R[j] 
                        j+=1    
            k+=1
          
        while i < len(L): 
            temp[k] = L[i] 
            i+=1
            k+=1
          
        while j < len(R): 
            temp[k] = R[j] 
            j+=1
            k+=1
```

# Function For Creating Posting

```python
def create_posting(posting_list):
 new_temp_list=[]
 new_term_index_list=[]
 term_index_values=posting_list[0]
 temp_index=[term_index_values[0],1,1,term_index_values[1],term_index_values[2]]
 new_temp_list=new_temp_list+temp_index
 temp_index.clear()
 length_list=len(posting_list)
 for i in range(1,length_list):
     length=len(new_temp_list)
     term_index_values=posting_list[i]
     if term_index_values[0]==new_temp_list[0]:
         new_temp_list[1]+=1
         if term_index_values[1]!=new_temp_list[length-2]:
             new_temp_list[2]+=1
         temp_index=[term_index_values[1],term_index_values[2]]
         new_temp_list=new_temp_list+temp_index
         temp_index.clear()
     else:
         new_term_index_list.append(new_temp_list.copy())
         new_temp_list.clear()
         temp_index=[term_index_values[0],1,1,term_index_values[1],term_index_values[2]]
         new_temp_list=new_temp_list+temp_index
         temp_index.clear()
 return new_term_index_list
```

# For Creating term_index.txt file With Delta Encoding

```python
f=open("term_index.txt","w+")
for key,val in term_index.items():
    f.write(str(key))
    f.write(' ')
    length=len(val)
    p_docid=val[2]
    p_term_pos=val[3]
    for i in range(4,length,2):
        n_docid=val[i]
        n_term_pos=val[i+1]
        val[i]=n_docid-p_docid
        if val[i]==0:
            val[i+1]=n_term_pos-p_term_pos
        p_docid=n_docid
        p_term_pos=n_term_pos
    k=1
    for v in val:
        f.write(str(v))
        if k==1 or k==2 or k%2==0:
            f.write(' ')
        else:
            f.write(',')
        k+=1
    f.write('\n')
f.close()
```

# For Retrieving term_index.txt file

```python
import re
new_term_index=dict()
f=open("term_index.txt","r",encoding='utf8', errors='ignore')
key=0
val=[]
for line in f.readlines():
    contents = line
    word=re.split(' |,',contents)
    word.pop(-1)
    key=int(word[0])
    val=[]
    for txt in word[1:]:
        val.append(int(txt))
    new_term_index[key]=val
    #print (key, val)
```
